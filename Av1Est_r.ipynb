Questão 1: Concentração de Mg versus Número de Registro

Obs: las controla a direção das legendas dos eixos (las= 1, legendas escritas sempre na horizontal, las=3, legendas sempre na vertical), sejam os números da escala ou o nome do eixo. 

Observar o fgl dentro da biblioteca MASS.

```
data.frame(MASS::fgl)
```

Verificação do nome das colunas do dataframe.

```
colnames(fgl)
```

---
Construção do gráfico Número de Registro versus Concentração de Mg com a utilização da bibliotteca ggplot para geração do gráfico.
---
theme_bw() para tornar a visualização do gráfico mais agradável.





```
x <- seq(1,214)
y <- fgl$Mg
graf1 <- ggplot(data = fgl, aes(x, y))
graf2 <- graf1 + geom_point() + geom_line() 
graf3 <- graf2 + labs(las = 1, title = "Número de Registro versus Concentranção de Mg", x = "Número de Registro", y = "Concentração de Mg")
graf4 <- graf3 + theme_bw()
graf4
```

Questão 2: Concentração de Ca versus Índice de Refração

Plotagem de gráfico Concentração de Ca versus Índice de Refração com o uso do las para tornar os rótulos horizontais em ambos os eixos.
Como resulatdo encontramos um gráfico de dispersão.

```
x <- fgl$Ca
y <- fgl$RI

plot(x,y, xlab="Concentração de Ca", ylab="Índice de Refração", main ="Concentração de Ca versus Índice de Refração", las=1, pch=20, col = "pink")
```

Resolução da 3° questão letras a, b e c, com uso de mfrow, pty = "s" para plotagem quadrada e transformação de y em fator.

```
x <- UScereal$mfr
y <- UScereal$shelf
par(mfrow = c(2,2))
plot(x,y, main="shelf como numérico", data= UScereal,xlab ="mfr", ylab = "shelf",las=1, pty = "s", col = "blue")
 
factshelf <- as.factor(y)
plot(x, factshelf, las=1, pty = "s", xlab= "mfr", ylab = " shelf factor", main = "shelf como fator")
```
